```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
```

```python
csv_files = [f for f in os.listdir() if f.endswith('.csv')]
dataframes = [pd.read_csv(file) for file in csv_files]
combined_df = pd.concat(dataframes, ignore_index=True)

# Define a function to apply color to each cell
def apply_color(value):
    color = plt.cm.RdYlGn(value)
    hex_color = mcolors.rgb2hex(color)
    return f'background-color: {hex_color}'

# Sort the DataFrame based on some metric; here I'm arbitrarily using 'low_precision'
#combined_df = combined_df[combined_df["method"]=="rougeL"]
combined_df = combined_df.loc[combined_df['mid_recall'].sort_values(ascending=True).index]


# Apply the coloring function
styled_df = combined_df.style.applymap(apply_color, subset=pd.IndexSlice[:, 'low_precision':])
styled_df
```
